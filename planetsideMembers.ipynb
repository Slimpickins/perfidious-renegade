```python
import pandas as pd 
import json
import requests
import time
import csv
from pandas import Series, DataFrame
```

```python
#Getting the member list from daybreak

call = 'https://census.daybreakgames.com/get/ps2:v2/outfit/?outfit_id=37530159341115681&c:resolve=member_character(name)'
resp = requests.get(url=call)
temp = resp.json()
#df = pd.read_json('https://census.daybreakgames.com/get/ps2:v2/outfit/?outfit_id=37530159341115681&c:resolve=member_character(name)')
```

```python
# Filtering to get only the members
# We have raw data if you want to get a new memberlist with current data start again here
members =  temp['outfit_list'][0]
data = members["members"][0:]
nameDF = DataFrame.from_dict(data)

#Expanding 'name' column dictionary to return only 'first_lower' name.
nameDF2 = nameDF['name'].apply(Series)
nameDF2.drop(nameDF2.columns[[0,2]], inplace=True, axis=1)
nameDF.drop('name', inplace=True, axis=1)
nameDF['name'] = nameDF2[[0]]

nameDF.head()
nameDF.to_csv('aodmems.csv', index=False)
```

```python
#Run this if you want only full-members. 'rank ordinal' values are 1-5, 5 being all members 1 being leader,
# so NCOs start at 3
nameDF = nameDF[nameDF['rank_ordinal'].astype('float') <5]
nameDF.head()
nameDF.to_csv('aodfullmems.csv', index=False)
```
